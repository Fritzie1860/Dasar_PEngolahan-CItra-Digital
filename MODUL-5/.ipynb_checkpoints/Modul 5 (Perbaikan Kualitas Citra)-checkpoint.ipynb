```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
mcitra = cv2.imread("rem_nangis.jpg",0)
plt.imshow(mcitra,"gray")
plt.title("Citra")
print(type(mcitra))
```

Lakukan perbaikan citra pada gambar di atas menggunakan window 3x3, 5x5 dan 7x7 dengan 
menggunakan proses filterinig median, mean, dan modus

# MEDIAN FILTERING

```python
def funcMed(matriks):
    bar,kol = matriks.shape
    n = bar*kol
    array = np.sort(matriks, axis=None)
    
    if n%2 == 0:
        a = array[n//2] # bagi dan bulatkan ke bilangan bulat terkecil berikutnya
        b = array[n//2 - 1]
        c = (a + b)/2
    else:
        c = array[n//2]
    
    return c

def med(citra, bar, kol): 
    result = np.zeros(citra.shape) 
    Bar, Kol = int((bar-1)/2), int((kol-1)/2)
    pad = np.pad(citra, ((Bar,bar-(Bar-1)),(Kol,kol-(Kol-1))), "constant")
    
    for i in range(len(citra)):
        for j in range(len(citra[i])):
            result[i,j] = funcMed(pad[i:i+bar, j:j+kol])
            
    return result
```

```python
plt.figure(figsize=(20,7)) 
plt.subplot(1,3,1) 
plt.imshow(med(mcitra, 3, 3), "gray")
plt.title("Median Wondow 3x3")
plt.subplot(1,3,2) 
plt.imshow(med(mcitra, 5, 5), "gray")
plt.title("Median Wondow 5x5")
plt.subplot(1,3,3) 
plt.imshow(med(mcitra, 7, 7), "gray")
plt.title("Median Wondow 7x7")
plt.show()
```

# MEAN FILTERING

```python
def funcMean(matriks):
    m = np.sum(matriks)
    bar, kol = matriks.shape
    
    return (m/(bar*kol))

def mean(citra, bar, kol): 
    result = np.zeros(citra.shape) 
    Bar, Kol = int((bar-1)/2), int((kol-1)/2)
    pad = np.pad(citra, ((Bar,bar-(Bar-1)),(Kol,kol-(Kol-1))), "constant")
    
    for i in range(len(citra)):
        for j in range(len(citra[i])):
            result[i,j] = funcMean(pad[i:i+bar, j:j+kol])
            
    return result
```

```python
plt.figure(figsize=(20,7)) 
plt.subplot(1,3,1) 
plt.imshow(mean(mcitra, 3, 3), "gray")
plt.title("Mean Wondow 3x3")
plt.subplot(1,3,2) 
plt.imshow(mean(mcitra, 5, 5), "gray")
plt.title("Mean Wondow 5x5")
plt.subplot(1,3,3) 
plt.imshow(mean(mcitra, 7, 7), "gray")
plt.title("Mean Wondow 7x7")
plt.show()
```

# MODUS FILTERING

```python
def funcModus(matriks):
    mat = np.sort (matriks, axis=None)
    indeks = np.zeros(np.max(mat)+1, dtype = int)
    
    for i in range (len(mat)):
        indeks[mat[i]] += 1
        
    maks = np.max(indeks)
    for j in range (len(indeks)):
        if (indeks[j] == maks):
            return j

def modus(citra, bar, kol): 
    result = np.zeros(citra.shape) 
    Bar, Kol = int((bar-1)/2), int((kol-1)/2)
    pad = np.pad(citra, ((Bar,bar-(Bar-1)),(Kol,kol-(Kol-1))), "constant")
    
    for i in range(len(citra)):
        for j in range(len(citra[i])):
            result[i,j] = funcModus(pad[i:i+bar, j:j+kol])
            
    return result
```

```python
plt.figure(figsize=(20,7)) 
plt.subplot(1,3,1) 
plt.imshow(modus(mcitra, 3, 3), "gray")
plt.title("Modus Wondow 3x3")
plt.subplot(1,3,2) 
plt.imshow(modus(mcitra, 5, 5), "gray")
plt.title("Modus Wondow 5x5")
plt.subplot(1,3,3) 
plt.imshow(modus(mcitra, 7, 7), "gray")
plt.title("Modus Wondow 7x7")
plt.show()
```
