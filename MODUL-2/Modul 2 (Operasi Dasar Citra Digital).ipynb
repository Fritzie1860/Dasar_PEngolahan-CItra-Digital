```python
# melakukan import library
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import os 
```

```python
gambar = cv2.imread("waifu.jpg") # mengimpor waifu.jpg dalam variabel gambar, default impor melalui cv2 adalah BGR
gbr = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY) # melakukan konversi warna BGR ke RGB
plt.imshow(gbr, cmap="gray") 
```

# 1. TRANSLATION

Translasi sejauh x = 50 dan y = 50

```python
x = 50
y = 50
trans = np.zeros([len(gbr)+x,len(gbr[0])+y])
for a in range (len(gbr)):
    for b in range (len(gbr[0])):
        trans [a+x][b+y] = gbr[a][b]
plt.imshow(trans, cmap="gray")
```

# 2. ROTATION

Rotasi dengan sudut 30 derajat

```python
rot = np.zeros([450,450])
_0 = np.deg2rad(30)
cos0 = np.cos(_0)
sin0 = np.sin(_0)
for x in range (len(gbr)):
    for y in range (len(gbr[0])):
        a=(int (cos0*(x-len(gbr)/2)-sin0*(y-len(gbr)/2)+len(gbr)/2))
        b=(int (sin0*(x-len(gbr)/2)+cos0*(y-len(gbr)/2)+len(gbr)/2)) 
        rot[a+100][b+100] = gbr[x][y]
plt.imshow(rot, cmap="gray")
```

# 3. REFLECTION

Pencerminan terhadap sumbu x dan sumbu y

```python
flip = np.zeros([len(gbr),len(gbr[0])])
for x in range (len(gbr)):
    for y in range (len(gbr)):
        flip [x][y]=gbr[-x][y]

print("Cermin Sumbu X")
plt.figure(figsize=(10,10)) 
plt.subplot(2,1,1) 
plt.imshow(gbr,cmap="gray")
plt.subplot(2,1,2) 
plt.imshow(flip,cmap="gray") 
```

```python
flip = np.zeros([len(gbr),len(gbr[0])])
for x in range (len(gbr)):
    for y in range (len(gbr)):
        flip [x][y]=gbr[x][-y]

print("Cermin Sumbu Y")
plt.figure(figsize=(10,10)) 
plt.subplot(1,2,1) 
plt.imshow(gbr,cmap="gray")
plt.subplot(1,2,2) 
plt.imshow(flip,cmap="gray") 
```

# 4. DILATATION

Dilatasi dengan faktor skala = 3

```python
zoom = 3
dilat = np.zeros([zoom*len(gbr),zoom*len(gbr[0])])
for x in range (len(gbr)):
    for y in range(len(gbr[0])):
        for i in range (x*zoom,x*zoom+zoom):
            for j in range (y*zoom,y*zoom+zoom):
                dilat[i,j]=gbr[x,y]
plt.imshow(dilat, cmap="gray")
print("size sebelum: ",gbr.size)
print("size sesudah: ",dilat.size)
```
