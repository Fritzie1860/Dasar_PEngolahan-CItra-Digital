# DEKOMPOSISI

1. Lakukan Dekomposisi Wavelet Level 1 pada citra di bawah ini menggunakan kernel dekomposisi Haar!

```python
# Melakukan import library
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Melakukan import citra
mcitra = cv2.imread("spek.jpeg",0)
np.resize(mcitra,(1000,1000))
plt.imshow(mcitra,"gray")
plt.title("Citra")
print(type(mcitra))
```

```python
# Inisialisasi filter
lo_d = np.array([0.7071067811865476, 0.7071067811865476])
hi_d = np.array([-0.7071067811865476, 0.7071067811865476])
lo_r = np.array([0.7071067811865476, 0.7071067811865476])
hi_r = np.array([0.7071067811865476, -0.7071067811865476])
```

Pusat filter (kernel) pada matriks 1x2 tersebut ada pada elemen (0,0). Untuk pendefinisian filter pada konvolusi kolom, perlu dilakukan transpose pada matriks filter. Variabel lo_d digunakan sebagai lowpass filter dan hi_d sebagai highpass filter.

```python
# Fungsi penambahan padding dinamis pada citra
def padding (citra, n_pad):
    n_pad = len(n_pad)-1
    padd = np.pad (citra , [0, n_pad], 'constant')
    
    return padd
```

Setelah mengetahui ukuran dari matriks filter, maka perlu dibuatkan padding pada matriks citra. Hal ini bertujuan agar seluruh elemen pada matriks citra dapat dikonvolusi seutuhnya. Setelah isi dari parameter citra dan n_pad telah berhasil dikirim ke dalam fungsi padding. Selanjutnya di dalamnya, dilakukan pemanggilan fungsi library numpy dalam variabel padd untuk pembuatan padding dengan parameter berisi array yaitu citra, pad_width yaitu [0, n_pad], dan mode constant untuk memerikan padding dengan value yang konstan. Yang mana n_pad merupakan panjang array dari filter dikurangi dengan 1. Fungsi ini akan mengembalikan nilai berupa isi dari variabel padd.

```python
# Fungsi konvolusi baris pada citra 
def konvoBar (padd, filter, citra):
    baris, kolom =  citra.shape
    konvo = np.zeros ((baris,kolom))
    for i in range (len(konvo)):
        for j in range (len(konvo[i])):
            for x in range (len(filter)):
                konvo[i][j] += (padd[i][j+x]*filter[x])

    return konvo
```

Fungsi konvoBar memiliki parameter berupa matriks yang telah diberi padding, matriks filternya, dan matriks citra itu sendiri. Matriks citra akan digunakan untuk mencari panjang dan lebar matriks yang dimanfaatlan untuk menyediakan matriks nol berbentuk mariks citra. Setelah memiliki matriks nol, dilakukan perulangan sebanyak panjang baris dari matriks nol yang sebelumnya. Kemudian selama perulangan itu terjadi, dilakukan perulangan sebanyak panjang kolom dari matriks nol. Di dalamnya terdapat proses perulangan yang akan berulang sebanyak panjang baris matriks filter. Sehingga di dalamnya, terdapat proses pengisian mariks nol dengan isi dari matriks nol yang sebelumnya ditambah elemen matriks citra (yang pada iterasi selanjutnya, indeks yang bertambah adalah indeks pada kolom) yang dikalikan dengan isi dari matriks filternya. Di akhir proses, fungsi ini akan mengembalikan nilai berupa matriks nol yang sudah melalui proses pengisian sebelumnya.

```python
# Fungsi konvolusi kolom pada citra 
def konvoKol (padd, filter, citra):
    baris, kolom =  citra.shape
    konvo = np.zeros ((baris,kolom))
    for i in range (len(konvo)):
        for j in range (len(konvo[i])):
            for x in range (len(filter)):
                konvo[i][j] += (padd[i+x][j]*filter[x])

    return konvo
```

Sama halnya dengan fungsi konvoBar, hanya saja pada proses pengisian darimariks nol dengan isi dari matriks nol yang sebelumnya ditambah elemen matriks citra (yang pada iterasi selanjutnya, indeks yang bertambah adalah indeks pada baris) yang dikalikan dengan isi dari matriks filternya.

```python
# Fungsi down sampling pada kolom ganjil
def downKol (mat_konvo):
    baris,kolom = mat_konvo.shape
    baris=int(baris)
    kolom=int(kolom/2)
    Genap = np.zeros((baris,kolom))   
    for i in range(baris):
        x = 0
        for j in range(kolom):
            Genap[i][j] = mat_konvo[i][x]
            x += 2
            
    return Genap
```

Fungsi downKol merupakan fungsi yang digunakan untuk melakukan down sampling terhadap kolom yang ganjil. Parameter dari fungsi ini adalah matriks yang telah melalui konvolusi. Panjang dari kolom pada matriks hasil konvolusi tersebut dibagi dengan 2. Hal ini bertujuan untuk menghasilkan matriks dalam variabel Genap dengan panjang baris yang tetap dan pajang kolom yang hanya tersisa setengahnya. Kemudian, dilakukan perulangan sebanyak panjang baris tersebut dan juga panjang kolom. Matriks dalam variabel Genap diisikan dengan matriks hasil konvolusi dengan indeks baris merujuk pada perulangan sarangnya, dan indeks kolom yang berisikan variabel yang akan selalu bertambah 2 selama perulangan pada kolom berlangsung. Di akhir proses yang ada, dilakukan pengembalian nilai pada variabel Genap yang berisi matriks baris utuh dan berkolom genap saja.

```python
# Fungsi down sampling pada baris ganjil
def downBar (mat_konvo):
    baris,kolom=mat_konvo.shape
    baris=int(baris/2)
    kolom=int(kolom)
    Genap = np.zeros((baris,kolom))
    x = 0

    for i in range(baris):
        for j in range(kolom):
            Genap[i][j] = mat_konvo[x][j]
        x += 2
        
    return Genap
```

Sama dengan proses yang ada pada fungsi downKol, hanya saja fungsi ini digunakan untuk melakukan down sampling terhadap baris ganjil sebuah matriks yang telah berisikan kolom genap saja. Sehingga di dalamnya, jumlah baris dibagi dengan 2. Kemudian pada proses pengisian matriks, indeks yang selalu bertambah 2 adalah indeks pada baris.

```python
# Fungsi proses dekomposisi
def dekomposisi (citra, f1, f2):
    pad = padding (citra, f1)
    konvolusi = konvoBar(pad, f1, citra)
    downsampling = downKol (konvolusi)
    padd = padding (downsampling, f2)
    konvolusii = konvoKol(padd,f2, downsampling)
    
    return downBar (konvolusii)
```

```python
# Inisialisasi pada tiap-tiap hasil dekomposisi
A = dekomposisi(mcitra, lo_d, lo_d)
dH = dekomposisi(mcitra, lo_d, hi_d)
dV = dekomposisi(mcitra, hi_d, lo_d)
dD = dekomposisi(mcitra, hi_d, hi_d)
```

```python
# Fungsi menampilkan 4 citra dalam skala citra semula
def lvl1 (LL, LH, HL, HH):
    baris_1 = np.concatenate ((LL, HL), axis = 1) # concatenate digunakan untuk menggabungkan LL & HL
    baris_2 = np.concatenate ((LH, HH), axis = 1)
    return np.concatenate ((baris_1, baris_2), axis=0)
```

```python
# Menampilkan 4 citra hasil dekomposisi dalam skala citra semula
hasil = lvl1 (A, dH, dV, dD)
plt.imshow(hasil,'gray')
```

```python
# Fungsi menampikan 4 citra dalam satu baris
def tampil(a, b, c, d):
    plt.figure(figsize=(15,15))
    plt.subplot(1,4,1)
    plt.imshow(a, cmap='gray')
    plt.subplot(1,4,2)
    plt.imshow(b, cmap='gray')
    plt.subplot(1,4,3)
    plt.imshow(c, cmap='gray')
    plt.subplot(1,4,4)
    plt.imshow(d, cmap='gray')
```

```python
# Menampilkan 4 citra hasil dekomposisi dalam satu baris
tampil (A, dH, dV, dD)
```

# REKONSTRUKSI

2. Lakukan Rekonstruksi Wavelet terhadap hasil jurnal nomor 1 menggunakan kernel rekonstruksi Haar

```python
# Fungsi rekonstruksi
def rekonstruksi(A, dV, dH, dD):
    baris, kolom = A.shape 
    k = kolom*2
    b = baris*2 
    img = np.zeros([k,b]) 
    LL, LH = np.zeros([k,b]), np.zeros([k,b]) 
    HL, HH = np.zeros([k,b]), np.zeros([k,b]) 
    
    # Memberikan ruang kosong pada matriks
    for i in range(0,b,2):
        for j in range(0,k,2):
            LL[i][j] = A[int(i/2)][int(j/2)]
            LH[i][j] = dV[int(i/2)][int(j/2)]
            HL[i][j] = dH[int(i/2)][int(j/2)]
            HH[i][j] = dD[int(i/2)][int(j/2)]
    
    # Mengisi baris yang kosong pada matriks citra
    for i in range(1,b,2):
        for j in range(k):
            LL[i][j] = LL[i-1][j]
            HL[i][j] = HL[i-1][j]
            LH[i][j] -= LH[i-1][j]
            HH[i][j] -= HH[i-1][j]
    
    # Mengisi kolom yang kosong pada matriks citra
    for i in range(b):
        for j in range(1,k,2):
            LL[i][j] = LL[i][j-1]
            HL[i][j] -= HL[i][j-1]
            LH[i][j] = LH[i][j-1]
            HH[i][j] -= HH[i][j-1]
    
    # Penggabungan LL, LH, HL, HH ke dalam satu matriks citra
    for i in range(b):
        for j in range(k):
            img[i][j] += (LL[i][j]+LH[i][j]+HL[i][j]+HH[i][j])
    return img
```

```python
hasil = rekonstruksi(A, dV, dH, dD) 
plt.imshow(hasil, cmap='gray')
```
