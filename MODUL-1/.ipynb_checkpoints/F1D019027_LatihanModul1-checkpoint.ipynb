# KONVOLUSI DAN NORMALISASI

```python
# import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimage
import os
```

```python
# mendefinisikan kernel
kernel = np.array([[0, 2, 0],
                   [-2, 5, -2],
                   [0, 2, 0]])
type(kernel)
print("Kernel: \n", kernel)
```

```python
# mengimpor citra
gambar = cv2.imread("bunga.jpg") 

# mengonversi color space citra menjadi GRAY 
gbr = cv2.cvtColor(gambar, cv2.COLOR_BGR2GRAY) 
```

```python
# ukuran citra
print("\nUkuran Citra: ", gbr.shape)

# matriks citra sebelum dan sesudah resizing
print ("\nCitra Asli: \n", gbr)
size = cv2.resize(gbr,(10,10))
print("\nSesudah Resizing: \n", size)

# citra yang akan dikonvolusi
print("\nCitra: \n", plt.imshow(gbr,cmap="gray"))
```

```python
# memberikan zero padding menggunakan library
padding = np.pad(gbr,[[1,1],[1,1]],'constant') 
print(padding[0:len(gbr)+2,0:len(gbr[0])+2])
```

```python
konvo = np.zeros([len(padding),len(padding[0])]) 
# operasi konvolusi dengan kernel
for i in range(1,len(padding)-1): 
    for j in range (1,len(padding[0])-1): 
        a = padding[i-1,j-1]*kernel [0,0] + padding[i-1,j]*kernel [0,1] + padding[i-1,j+1]*kernel [0,2]
        b = padding[i,j-1]*kernel [1,0] + padding[i,j]*kernel [1,1] + padding[i,j+1]*kernel [1,2]
        c = padding[i+1,j-1]*kernel [2,0] + padding[i+1,j]*kernel [2,1] + padding[i+1,j+1]*kernel [2,2]
        konvo[i-1,j-1] = a+b+c
```

```python
# menampilkan citra semula dengan hasil konvolusi 
plt.figure(figsize=(10,10)) 
plt.subplot(1,2,1) 
plt.imshow(gbr,cmap="gray")
plt.subplot(1,2,2) 
plt.imshow(konvo,cmap="gray")
```

[RUMUS NORMALISASI]
norm(x) = ((x - min)*max)/rentang
rentang = max-min

```python
# Normalisasi

# menentukan intensitas maksimum dan minimum dari citra
maks = np.max(konvo)
print("Max: ", maks)

mins = np.min(konvo)
print("Min:", mins)

rtg = maks-mins
print("Rentang: ", rtg)

normal = np.zeros([len(konvo)-1,len(konvo[0])-1])
for i in range(len(normal)): 
    for j in range (len(normal[0])): 
        normal[i,j]= round(((konvo[i,j]-mins)/rtg)*255)
        
print("\nMatriks Citra:\n", normal)
```

```python
# menampilkan citra semula dengan hasil konvolusi 
plt.figure(figsize=(15,15)) 
plt.subplot(1,3,1) 
plt.imshow(gbr,cmap="gray")
plt.subplot(1,3,2) 
plt.imshow(konvo,cmap="gray")
plt.subplot(1,3,3) 
plt.imshow(normal,cmap="gray")
```
